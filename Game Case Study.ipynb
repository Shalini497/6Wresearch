*** Difference between normal and student t distribution explain it using 'potential' column. how to show this in python ***

```python
from scipy.stats import norm, t
import numpy as np

# Assuming 'df' is the DataFrame containing the 'potential' column
potential_data = df['Potential'].dropna()  # Remove any missing values

# Fit a normal distribution to the 'potential' data
mu_normal, std_normal = norm.fit(potential_data)

# Fit a Student's t-distribution to the 'potential' data
df_t, mu_t, std_t = t.fit(potential_data)

# Generate values for x-axis (potential values)
x = np.linspace(min(potential_data), max(potential_data), 100)

# Calculate the PDF for both distributions
pdf_normal = norm.pdf(x, mu_normal, std_normal)
pdf_t = t.pdf(x, df_t, mu_t, std_t)

# Plot the PDFs
plt.figure(figsize=(8, 6))
plt.plot(x, pdf_normal, label='Normal Distribution', color='green')
plt.plot(x, pdf_t, label="Student's t-distribution", color='red')
plt.hist(potential_data, density=True, bins=20, alpha=0.6, color='gray', label='Histogram')

plt.xlabel('Potential')
plt.ylabel('Density')
plt.title('Normal vs. Student\'s t-distribution for Potential')
plt.legend()
plt.grid(True)
plt.show()

```

*** Difference between normal and standard normal distribution explain it using 'potential' column.	***

```python
# Assuming 'df' is the DataFrame containing the 'potential' column
potential_data = df['Potential'].dropna()  # Remove any missing values

# Calculate the mean and standard deviation of the 'potential' data
mean_potential = potential_data.mean()
std_potential = potential_data.std()

# Generate values for x-axis (potential values)
x = np.linspace(mean_potential - 3 * std_potential, mean_potential + 3 * std_potential, 100)

# Calculate the probability density function (PDF) for standard normal distribution
pdf_standard_normal = stats.norm.pdf(x, loc=mean_potential, scale=std_potential)

# Plot the histogram for 'potential' data
plt.figure(figsize=(8, 6))
plt.hist(potential_data, density=True, bins=20, alpha=0.6, color='gray', label='Potential Data')

# Plot the PDF for the standard normal distribution
plt.plot(x, pdf_standard_normal, label='Standard Normal Distribution', color='blue')

plt.xlabel('Potential')
plt.ylabel('Density')
plt.title('Potential Data vs. Standard Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```

*** find the 95%, 90%, and 99%, confidence interval for 'Potential','wage','weight' column. ***

```python
# Assuming 'df' is the DataFrame containing the data

# Columns for which we want to calculate confidence intervals
columns_to_analyze = ['Potential', 'Wage', 'Weight']

# Confidence levels
confidence_levels = [0.95, 0.90, 0.99]

# Dictionary to store the confidence intervals
confidence_intervals = {}

# Loop through each column and calculate the confidence intervals
for column in columns_to_analyze:
    data_column = pd.to_numeric(df[column], errors='coerce').dropna()  # Remove non-numeric values
    sample_size = len(data_column)
    
    if sample_size == 0:
        print(f"No numeric values found in column '{column}'.")
        continue
    
    sample_mean = data_column.mean()
    sample_std = data_column.std(ddof=1)  # Use Bessel's correction for sample standard deviation
    
    for confidence_level in confidence_levels:
        # Calculate the critical value for the confidence level
        critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)
        
        # Calculate the margin of error
        margin_of_error = critical_value * (sample_std / (sample_size ** 0.5))
        
        # Calculate the confidence interval
        lower_bound = sample_mean - margin_of_error
        upper_bound = sample_mean + margin_of_error
        
        confidence_intervals[(column, confidence_level)] = (lower_bound, upper_bound)

# Display the confidence intervals
for column, confidence_level in confidence_intervals:
    lower_bound, upper_bound = confidence_intervals[(column, confidence_level)]
    print(f"{column} {confidence_level * 100}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

*** find Central theorem ***

```python
# Assuming 'df' is the DataFrame containing the data

# Extract the 'Potential' column from the DataFrame
potential_data = df['Potential'].dropna()  # Remove any missing values

# Number of samples to take and sample size per sample
num_samples = 1000
sample_size = 30

# List to store the sample means
sample_means = []

# Generate multiple samples and calculate the means
for _ in range(num_samples):
    sample = np.random.choice(potential_data, size=sample_size, replace=True)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Plot the histogram of the sample means
plt.figure(figsize=(8, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='gray', label='Sample Means')

# Calculate the mean and standard deviation of the 'Potential' data
population_mean = potential_data.mean()
population_std = potential_data.std()

# Calculate the standard deviation of the sample means
std_sample_means = np.std(sample_means, ddof=1)

# Calculate the theoretical standard deviation of the sample means using the Central Limit Theorem
clt_std = population_std / (sample_size ** 0.5)

# Generate values for x-axis (potential values)
x = np.linspace(min(sample_means), max(sample_means), 100)

# Calculate the probability density function (PDF) for the normal distribution
pdf_normal = (1 / (clt_std * (2 * np.pi) ** 0.5)) * np.exp(-0.5 * ((x - population_mean) / clt_std) ** 2)

# Plot the PDF for the normal distribution
plt.plot(x, pdf_normal, label='Normal Distribution', color='blue')

plt.xlabel('Sample Mean (Potential)')
plt.ylabel('Density')
plt.title('Central Limit Theorem - Sample Means vs. Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
